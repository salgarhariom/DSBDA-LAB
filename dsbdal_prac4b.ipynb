```python
x=data.drop(['price','zipcode'], axis=1)
```

```python
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
model=lm.fit(xtrain,ytrain)
```

```python
model
```

```python
ytrain_pred=lm.predict(xtrain)
ytest_pred=lm.predict(xtest)
```

```python

df=pd.DataFrame(ytrain_pred,ytrain)
df=pd.DataFrame(ytest_pred,ytest)
```

```python
from sklearn.metrics import mean_squared_error , r2_score
mse = mean_squared_error(ytest, ytest_pred)
print(mse)
```

```python
mse = mean_squared_error(ytrain_pred,ytrain)
mse
```

```python
plot.scatter(ytrain, ytrain_pred ,c='blue',marker='o',label='Training data')
plot.scatter(ytest,ytest_pred, c='lightgreen',marker='s',label='Test data')
plot.xlabel('True values')
plot.ylabel('Predicted')
plot.title("True value vs Predicted value")
plot.legend(loc= 'upper left')
plot.plot()
plot.show()
```
